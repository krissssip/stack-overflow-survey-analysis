```python
#Завдання 1. Підрахунок загальної кількості респондентів

num_respondents = df['ResponseId'].nunique()
print(f"Кількість респондентів в опитуванні: ",num_respondents)
```

```python
#Завдання 2. Аналіз повноти відповідей респондентів

questions = set(df_schema['qname'])
common_questions = questions.intersection(df.columns)
respondents_full = df[list(common_questions)].dropna().shape[0]
print(f"Кількість респондентів, які відповіли на всі запитання з опитування: ",respondents_full)
```

```python
#Завдання 3. Статистичний аналіз досвіду респондентів

print("Міри центральної тенденції: ")
print(f"Mean: ", df['WorkExp'].mean())
print(f"Median: ", df['WorkExp'].median())
print(f"Mode: ", df['WorkExp'].mode()[0])
```

```python
#Завдання 4. Аналіз віддаленої роботи
print(f'Всі можливі відповіді: ',df['RemoteWork'].unique())
```

```python
remote_employees = (df['RemoteWork']== 'Remote').sum()
print(f'Kількість респондентів, які працюють віддаленo: ',remote_employees)
```

```python
#Завдання 5. Визначення популярності Python

python_users = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

count_python = python_users.sum()
total = df.shape[0]

percentage = (count_python / total) * 100

print(f"Відсоток респондентів, які програмують на Python: {percentage:.2f}%")
```

```python
#Завдання 6. Аналіз шляхів навчання програмуванню - LearnCode

course_users = df['LearnCode'].str.contains('Online Courses', na=False)

count_course = course_users.sum()


print(f"Кількість респондентів, які навчалися програмувати через онлайн курси: ", count_course)

```

```python
#Завдання 7. Географічний аналіз компенсації Python-розробників

df_python = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
df_python = df_python.dropna(subset=['ConvertedCompYearly'])
df_python_grouped = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean','median'])
df_python_grouped
```

```python
#Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів
df_sort = df.dropna(subset=['ConvertedCompYearly'])
df_sort = df_sort.sort_values(by= 'ConvertedCompYearly', ascending=False)
df_top_5 = df_sort.head()
df_top_5[['EdLevel','ConvertedCompYearly']]
```
